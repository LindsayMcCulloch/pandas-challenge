# PyCity Schools Analysis

* Schools with a smaller number of students have better overall grades than schools with more students
* Overall , student budgets of $585-630 range produce better results 
---

```python
# Dependencies and Setup
import pandas as pd
import numpy as np
# File to Loead (Remember to Change These)
SchoolDataCSV = "Resources/schools_complete.csv"
StudentDataCSV = "Resources/students_complete.csv"
# Read School and Student Data File and store into Pandas DataFrames
SchoolData = pd.read_csv(SchoolDataCSV)
StudentData = pd.read_csv(StudentDataCSV)
# Combine the data into a single dataset.
CombinedData = pd.merge(StudentData, SchoolData, how="left", on=[
                                "school_name", "school_name"])
```

## Local Government Area Summarys AND reading)

```python
#Calculate the total number of unique schools (2 points)
#Calculate the total number of schools
AreaSchoolsRows = CombinedData["school_name"].unique()
AreaSchoolsCount = len(AreaSchoolsRows)
print(f"The total number of schools is {AreaSchoolsCount}")
```

```python
#Calculate the total number of students (2 points)
#Calculate the total number of students
AreaStudentsCount = StudentData ["student_name"].count()
print(f"The total number of students is {AreaStudentsCount}")
```

```python
#Calculate the total budget (2 points)
#Calculate the total budget
AreaTotalBudget = SchoolData["budget"].sum()
print(f"The total budget is {AreaTotalBudget}")
```

```python
#Calculate the average (mean) maths score (2 points)
#Calculate the average maths score
AreaAverageMaths = CombinedData["maths_score"].mean()
print(f"The average maths score is {AreaAverageMaths}")
```

```python
#Calculate the average (mean) reading score (2 points)
#Calculate the average reading score
AreaAverageReading = CombinedData["reading_score"].mean()
print(f"The average reading score is {AreaAverageReading}")
```

```python
#Use the code provided to calculate the percentage of students who passed maths (2 points)
#Calculate the percentage of students with a passing maths score (50 or greater)
AreaPassingMathScore = CombinedData.loc[CombinedData["maths_score"] >= 50,:]
AreaPassingMath = len(AreaPassingMathScore) / len(CombinedData["maths_score"])
AreaPassingMath = AreaPassingMath * 100
print(f"The percentage of students with a passing maths score is {AreaPassingMath}%")
```

```python
#Calculate the percentage of students who passed reading (2 points)
#Calculate the percentage of students with a passing reading score (50 or greater)
AreaPassingReadingScore = CombinedData.loc[CombinedData["reading_score"] >= 50,:]
AreaPassingReading = len(AreaPassingReadingScore) / len(CombinedData["reading_score"])
AreaPassingReading = AreaPassingReading * 100
print(f"The percentage of students with a passing reading score is {AreaPassingReading}%")
```

```python
#Use the code provided to calculate the percentage of students that passed both maths and reading (2 points)
#Calculate the percentage of students who passed maths **and** reading (% Overall Passing)
AreaPassingMathTotal = len(AreaPassingMathScore)
AreaPassingReadingTotal = len(AreaPassingReadingScore)
AreaPassingTotal = AreaPassingReadingTotal + AreaPassingMathTotal
AreaMathTotal = len(CombinedData["maths_score"])
AreaReadingTotal = len(CombinedData["reading_score"])
AreaTotal = AreaReadingTotal + AreaMathTotal
AreaCombinedTotal = AreaPassingTotal / AreaTotal
AreaCombinedTotal = AreaCombinedTotal * 100
print(f"The percentage of students with a passing reading score and a passing math score is {AreaCombinedTotal}%")
```

```python
#Create a new DataFrame for the above calculations called area_summary (4 points)
#Create a dataframe to hold the above results
area_summary = pd.DataFrame({
    "Total Number Of Schools": [AreaSchoolsCount],
    "Total Number Of Students": [AreaStudentsCount],
    "Total Budget": [AreaTotalBudget],
    "Average Maths Score": [AreaAverageMaths],
    "Average Reading Score": [AreaAverageReading],
    "% Passing Maths": [AreaPassingMath],
    "% Passing Reading": [AreaPassingReading],
    "% Overall Passing": [AreaCombinedTotal]
    })
```

```python
#Formatting and results
area_summary["Total Budget"] = area_summary["Total Budget"].map("${:,}".format)
area_summary["Total Number Of Students"] = area_summary["Total Number Of Students"].map("{:,}".format)
area_summary["Average Maths Score"] = area_summary["Average Maths Score"].map("{:.2f}%".format)
area_summary["Average Reading Score"] = area_summary["Average Reading Score"].map("{:.2f}%".format)
area_summary["% Passing Maths"] = area_summary["% Passing Maths"].map("{:.2f}%".format)
area_summary["% Passing Reading"] = area_summary["% Passing Reading"].map("{:.2f}%".format)
area_summary["% Overall Passing"] = area_summary["% Overall Passing"].map("{:.2f}%".format)
AreaSummary = area_summary
AreaSummary
```

```python
#Per student budget
SchoolStudentBudget = pd.merge(SchoolTotalBudget, SchoolTotalStudents, on="school_name")
SchoolStudentBudgetTotal = SchoolStudentBudget['budget'] / SchoolStudentBudget['Student ID']
SchoolStudentBudgetTotal
```

```python
#Calculate the number of schools with maths scores of 50 or higher (2 points)
#Passing maths
SchoolPassingMath = CombinedData.loc[CombinedData["maths_score"] >= 50,:].groupby(["school_name"])["maths_score"].count()
SchoolPassingMathTotal = SchoolPassingMath / SchoolTotalStudents * 100
SchoolPassingMathTotal
```

```python
#Calculate the number of schools with reading scores of 50 or higher (2 points)
#Passing reading
SchoolPassingReading = CombinedData.loc[CombinedData["reading_score"] >= 50,:].groupby(["school_name"])["reading_score"].count()
SchoolPassingReadingTotals = SchoolPassingReading / SchoolTotalStudents * 100
SchoolPassingReadingTotals
```

```python
#Use the provided code to calculate the schools that passed both maths and reading with scores of 50 or higher (2 points)
#Overall passing
SchoolPassingTotal = (CombinedData.loc[(CombinedData["reading_score"] >= 50) & (CombinedData["maths_score"] >= 50)].groupby(["school_name"])["Student ID"].count())
SchoolTotal = SchoolPassingTotal / SchoolTotalStudents * 100
SchoolTotal
```

## Scores by School Spending

```python
# Establish the bins
SpendingRange = [0, 585, 630, 645, 680]
GroupSpending = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
#Use pd.cut with the provided code to bin the data by the spending ranges (2 points)
#Use the code provided to calculate the averages (1 points)
#Create the spending_summary DataFrame using the binned and averaged spending data (2 points)
#bin data and create new df
spending_summary = pd.DataFrame({
    "School Type": SchoolType,
    "Total Students": SchoolTotalStudents,
    "Total Budget": SchoolTotalBudget,
    "Per Student Budget": SchoolStudentBudgetTotal,
    "Average Maths Scores": SchoolAverageMaths,
    "Average Reading Scores": SchoolAverageReading,
    "% Passing Maths": SchoolPassingMathTotal,
    "% Passing Reading": SchoolPassingReadingTotals,
    "% Overall Passing": SchoolTotal
    })
```

```python
# Select only numeric columns for calculating the mean
numeric_columns = spending_summary.select_dtypes(include=[np.number])

# Create the "Per Student Budget Ranges" column
numeric_columns["Per Student Budget Ranges"] = pd.cut(spending_summary["Per Student Budget"], SpendingRange, labels=GroupSpending)

# Group by "Per Student Budget Ranges" and calculate the mean for numeric columns
summary_df2_group = numeric_columns.groupby(["Per Student Budget Ranges"], observed=False)
summary_df2_mean = summary_df2_group.mean()

# Create SpendingSummary by dropping non-numeric columns
SpendingSummary = summary_df2_mean.drop(['Total Students', 'Total Budget', 'Per Student Budget'], axis=1)
SpendingSummary
```

```python
SpendingSummary["Average Maths Scores"] = SpendingSummary["Average Maths Scores"].map("{:.2f}%".format)
SpendingSummary["Average Reading Scores"] = SpendingSummary["Average Reading Scores"].map("{:.2f}%".format)
SpendingSummary["% Passing Maths"] = SpendingSummary["% Passing Maths"].map("{:.2f}%".format)
SpendingSummary["% Passing Reading"] = SpendingSummary["% Passing Reading"].map("{:.2f}%".format)
SpendingSummary["% Overall Passing"] = SpendingSummary["% Overall Passing"].map("{:.2f}%".format)
SpendingSummary
```
